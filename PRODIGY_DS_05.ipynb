```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Weather': ['Clear', 'Rain', 'Fog', 'Clear', 'Rain', 'Fog', 'Clear'],
    'Road_Condition': ['Dry', 'Wet', 'Wet', 'Dry', 'Wet', 'Wet', 'Dry'],
    'Accidents': [20, 35, 15, 25, 40, 10, 30],
    'Hour': [8, 18, 22, 14, 19, 23, 10]
}

df = pd.DataFrame(data)

print(df)
```

```python
sns.barplot(x='Weather', y='Accidents', data=df)

plt.title("Accidents by Weather Condition")

plt.show()
```

```python
sns.barplot(x='Road_Condition', y='Accidents', data=df)

plt.title("Accidents by Road Condition")

plt.show()
```

```python
sns.lineplot(x='Hour', y='Accidents', data=df)

plt.title("Accidents by Hour")

plt.show()
```
